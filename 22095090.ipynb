```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

plt.style.use('ggplot')

df = pd.read_csv('WHO-COVID-19-global-data.csv')
```

```python
df['Date_reported'] = pd.to_datetime(df['Date_reported'])
global_new_cases = df.groupby('Date_reported')['New_cases'].sum()
top_countries = df.groupby('Country')[['Cumulative_cases', 'Cumulative_deaths']].max()
top_countries = top_countries.sort_values(by='Cumulative_cases', ascending=False).head(5)
cases_by_region = df.groupby('WHO_region')['Cumulative_cases'].max()
global_deaths = df.groupby('Date_reported')['Cumulative_deaths'].sum()
```

```python
plt.figure(figsize=(16, 14), constrained_layout=True)

gs = plt.GridSpec(3, 2, figure=plt.gcf())

plt.subplot(gs[0, 0])
plt.plot(global_new_cases.index, global_new_cases.values, color='blue', linewidth=2)
plt.title('Global Trend of New COVID-19 Cases', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Number of New Cases')
plt.grid(True)

plt.subplot(gs[0, 1])
top_countries.plot(kind='bar', ax=plt.gca())
plt.title('Top 5 Countries by Cumulative COVID-19 Cases and Deaths', fontsize=14, fontweight='bold')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Legend')

plt.subplot(gs[1, 0])
plt.pie(cases_by_region, labels=cases_by_region.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of COVID-19 Cases by WHO Region', fontsize=14, fontweight='bold')

plt.subplot(gs[1, 1])
plt.fill_between(global_deaths.index, global_deaths.values, color='red', alpha=0.5)
plt.title('Global Trend of Cumulative COVID-19 Deaths', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Cumulative Deaths')
plt.grid(True)

plt.suptitle("Global COVID-19 Pandemic Overview", fontsize=16, fontweight='bold', fontname='Arial')

student_id = "22095090"
student_name = "Sana"
plt.figtext(0.9, 0.98, f'Name: {student_name}, Student ID: {student_id}', ha='right', fontsize=16, fontweight='bold', fontname='Arial')

plt.subplot(gs[2, :])
plt.text(0.5, 0.95, "Description", ha='center', va='center', fontsize=14, fontweight='bold', fontname='Arial')
plt.axis('off')
description_text = """
The infographic "Global COVID Pandemic Overview" gives a nuanced point of view on the pandemic's qualities.
It is beyond question feasible for viewers to see the value in significant models and information because of the movement of instructive pictures that highlight significant parts of the issue in this infographic.
The chief chart shows the number of instances of global COVID. By delineating the instabilities of new sicknesses, it gives a realistic portrayal of the movement of the plague.
The going with sensible reference diagram delineates the five nations that the COVID pandemic has caused a stir in.
Determine the seriousness of the pandemic by leading an examination of the quantity of cases and fatalities that have happened in these nations.
The third point is that the WHO has an outline of COVID occasions.
This guide might help watchers grasp the geographical spread of the pandemic since it delineates areas for which illness rates are estimated in successive requests.
The last reasonable, which is an overview of the particular area, shows the global COVID loss of life. It uncovers the cruel truth of the passings that will happen in the future because of the pandemic.
It is beyond a shadow of a doubt feasible for viewers to understand the infographic because of the way that it contains sensible checks, portrayals, and succinct clarifications.
The discernments are difficultly determined to relate the Global COVID episode to satisfy the targets of the assignment.
In conclusion, the infographic named "Global COVID Pandemic Overview" isn't just outwardly engaging, but also instructive, and it is chipping away at pandemic data. To understand the pandemic that is clearing the world, it is fundamental.
"""
plt.text(0.5, 0.5, description_text, ha='center', va='center', fontsize=12, fontname='Arial', wrap=True)
```
